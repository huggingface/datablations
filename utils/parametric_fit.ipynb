```python
### Adapted from https://github.com/bigcode-project/bigcode-analysis/blob/91daccbc13cd972b7ce9627e6e1890c9affbf3cc/data_analysis/notebooks/ScalingLaws.ipynb ###
import torch
import numpy as np

def loss(inp, params, tie):
    if tie:
        a, b, e, alpha, beta = params[0], params[1], params[2], params[3], params[3]
    else:
        a, b, e, alpha, beta = params[0], params[1], params[2], params[3], params[4]

    pre_lse = torch.stack([a - alpha*torch.log(inp[:, 0]), b - beta*torch.log(inp[:, 1]), e.expand((inp.shape[0]))])
    post_lse = torch.logsumexp(pre_lse, dim=0)
    huber_loss = torch.nn.functional.huber_loss(post_lse, torch.log(inp[:, 2]), delta=1e-3, reduction='none')
    return huber_loss.sum()

def minimize_loss(inp, init_params=[6, 6, -1, 0.28, 0.28], steps=50, fixed=False, fix_e=False, tie=False):
    params = torch.nn.Parameter(data=torch.Tensor(init_params))
    if tie:
        params = torch.nn.Parameter(data=torch.Tensor(init_params[:-1]))

    lbfgs = torch.optim.LBFGS([params],
                    lr=1e-1,
                    history_size=10,
                    max_iter=20,
                    line_search_fn="strong_wolfe")

    def closure():
        lbfgs.zero_grad()
        l = loss(inp, params, tie=tie)

        l.backward()
        if fixed:
            # Do not backprop on fixed params
            # https://discuss.pytorch.org/t/set-a-portion-of-a-tensor-to-have-requires-grad-false/157856
            params.grad[:3] = 0
        elif fix_e:
            params.grad[2] = 0
        return l

    history_lbfgs = []
    for i in range(steps):
        l = lbfgs.step(closure)
    return l, params
```

```python
### Adapted from https://github.com/bigcode-project/bigcode-analysis/blob/91daccbc13cd972b7ce9627e6e1890c9affbf3cc/data_analysis/notebooks/ScalingLaws.ipynb ###
import torch
import numpy as np

def loss(inp, params):
    a, b, e, alpha, beta, ep_star, n_star = params

    pre_lse = torch.stack([
        a - alpha * torch.log(inp[:, 0] + (inp[:, 0] * n_star * (1 - torch.exp(-1*inp[:, 3]/n_star)))),
        b - beta * torch.log(inp[:, 1] + (inp[:, 1] * ep_star * (1 - torch.exp(-1*inp[:, 2]/ep_star)))),
        e.expand((inp.shape[0]))
    ])

    post_lse = torch.logsumexp(pre_lse, dim=0)
    huber_loss = torch.nn.functional.huber_loss(post_lse, torch.log(inp[:, 4]), delta=1e-3, reduction='none')

    return huber_loss.sum()

def minimize_loss(inp, init_params=[6, 6, 6, -1, 0.28, 0.32, 0.3], steps=50, fixed=False, fixed_except_b=False):
    params = torch.nn.Parameter(data=torch.Tensor(init_params))

    lbfgs = torch.optim.LBFGS([params],
                    lr=1e-1,
                    history_size=10,
                    max_iter=20,
                    line_search_fn="strong_wolfe")

    def closure():
        lbfgs.zero_grad()
        l = loss(inp, params)

        l.backward()
        if fixed:
            # Do not backprop on fixed params
            # https://discuss.pytorch.org/t/set-a-portion-of-a-tensor-to-have-requires-grad-false/157856
            params.grad[:5] = 0
        return l

    history_lbfgs = []
    for i in range(steps):
        l = lbfgs.step(closure)
    return l, params
```

```python
### Adapted from https://github.com/bigcode-project/bigcode-analysis/blob/91daccbc13cd972b7ce9627e6e1890c9affbf3cc/data_analysis/notebooks/ScalingLaws.ipynb ###
import torch
import numpy as np

def loss(inp, params):
    a, b, e, alpha, beta, ep_star, n_star = params

    """Feed data term into optimal N

    #a, b, e, alpha, beta, ep_star, n_star, x = params

    ts = inp[:, 1] + (inp[:, 1] * ep_star * (1 - torch.exp(-1*inp[:, 2]/ep_star)))
    ns = torch.minimum(optimal_N(D_to_C(ts*0.1)), inp[:, 0])

    ns_ratio = torch.maximum((inp[:, 0] / ns) - 1, torch.tensor(0))

    #ns_ratio = torch.maximum(torch.log(inp[:, 0]/ns), torch.tensor(0))

    pre_lse = torch.stack([
        a - alpha * torch.log(ns + (ns * n_star * (1 - torch.exp(-1*ns_ratio/n_star)))),
        b - beta * torch.log(ts),
        e.expand((inp.shape[0]))
    ])
    """

    ### Normal ###

    #pre_lse = torch.stack([
    #    a - alpha * torch.log(inp[:, 0] * ((1 - (1 - n_star)**inp[:, 3])/n_star)),
    #    b - beta * torch.log(inp[:, 1] * ((1 - (1 - ep_star)**inp[:, 2])/ep_star)),
    #    e.expand((inp.shape[0]))
    #])

    pre_lse = torch.stack([
        a - alpha * (torch.log(inp[:, 0]) + torch.log(1 - ((1 - n_star)**inp[:, 3])) - torch.log(n_star)),
        b - beta * (torch.log(inp[:, 1]) + torch.log(1 - ((1 - ep_star)**inp[:, 2])) - torch.log(ep_star)),
        e.expand((inp.shape[0]))
    ])

    #pre_lse = torch.stack([
    #    a - alpha * torch.log(inp[:, 0] + (inp[:, 0] * n_star * (1 - torch.exp(-1*inp[:, 3]/n_star)))),
    #    b - beta * torch.log(inp[:, 1] + (inp[:, 1] * ep_star * (1 - torch.exp(-1*inp[:, 2]/ep_star)))),
    #    e.expand((inp.shape[0]))
    #])

    ###

    """Special activation function

    # Learn the parameters x & y or fix them
    #a, b, e, alpha, beta, ep_star, n_star, y, z = params
    y = 0.99
    z = 0
    def oneminusexp(x): return 1 - torch.exp(-x)
    ACT_FN = lambda x: torch.where(x < z, x, z + oneminusexp(((x-z)*(1-z))**y)**(1/y)/(1-z))

    pre_lse = torch.stack([
      a - alpha * torch.log(inp[:, 0] + (inp[:, 0] * n_star * ACT_FN(inp[:, 3]/n_star))),
      b - beta * torch.log(inp[:, 1] + (inp[:, 1] * ep_star * ACT_FN(inp[:, 2]/ep_star))),
      e.expand((inp.shape[0]))
    ])

    """


    post_lse = torch.logsumexp(pre_lse, dim=0)
    huber_loss = torch.nn.functional.huber_loss(post_lse, torch.log(inp[:, 4]), delta=1e-3, reduction='none')

    return huber_loss.sum()

def minimize_loss(inp, init_params=[6, 6, 6, -1, 0.28, 0.32, 0.3], steps=50, fixed=False, fixed_except_b=False):
    params = torch.nn.Parameter(data=torch.Tensor(init_params))

    lbfgs = torch.optim.LBFGS([params],
                    lr=1e-1,
                    history_size=10,
                    max_iter=20,
                    line_search_fn="strong_wolfe")

    def closure():
        lbfgs.zero_grad()
        l = loss(inp, params)

        l.backward()
        if fixed:
            # Do not backprop on fixed params
            # https://discuss.pytorch.org/t/set-a-portion-of-a-tensor-to-have-requires-grad-false/157856
            params.grad[:5] = 0
        return l

    history_lbfgs = []
    for i in range(steps):
        l = lbfgs.step(closure)
    return l, params
```
